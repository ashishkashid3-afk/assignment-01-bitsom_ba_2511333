```python
sentences= [
# Cricket
             "The batsman Virat Kohli scored centuary in the final match",
             "The bowler Mitch Starc deleiver the perfect yorkers, who got most of the wickets in match",
             "The Indian team won the cricket match by 4 wickets",
             "The Virat Kohli Judged as Man of the match & Mitch got Man of the series",
# Cooking
             "The cook prepared delicious Mutton Biryani",
             "The receipe include fresh aroma of Indian spices",
             "Making a perfect cocktail required precision",
#Cybersecurity
             "Firewalls prevent unauthorised network acess",
             "Strong password is good to prevent cyber attacks",
             "Encryption ensures data privacy and security"
]
```

```python
from sentence_transformers import SentenceTransformer
model=SentenceTransformer('all-MiniLM-L6-v2')
embeddings=model.encode(sentences)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt

similarity_matrix=cosine_similarity(embeddings)

plt.figure(figsize=(8,6))
sns.heatmap(similarity_matrix, annot=True, cmap='coolwarm')
plt.title("Cosine_similarity_matrix")
plt.show()
```

```python
query = "The bowler took three wickets in one over"
query_embedding = model.encode([query])
similarities = cosine_similarity(query_embedding,embeddings)[0]
top_indices = similarities.argsort()[-2:][::-1]

for i in top_indices:
  print("Sentence:", sentences[i])
  print("Similarity Score:", similarities[i])
  print()
```
